`````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
`````

`````python
# Dataset artificial
np.random.seed(29)
n = 500
data = pd.DataFrame({
    'categoria': np.random.choice(['A', 'B', 'C'], size=n),
    'valor1': np.random.normal(50, 10, size=n),
    'valor2': np.random.normal(30, 5, size=n),
    'valor3': np.random.exponential(1.0, size=n),
    'valor4': np.random.randint(0, 100, size=n)
})
`````

# Visualización 1: Violin plot (distribución y densidad)

`````python
plt.figure(figsize=(8, 4))
sns.violinplot(x='categoria', y='valor1', data=data)
plt.title('Distribución de valor1 por categoría', fontsize=16, fontweight="bold")
plt.show()
`````

**Pregunta**

¿Qué diferencia se observa entre las categorías A, B y C en el violin plot de valor1?

¿Alguna categoría presenta una mayor dispersión de valores?

# Visualización 2: Boxen plot (útil para datos con colas largas)

`````python
plt.figure(figsize=(8, 4))
sns.boxenplot(x='categoria', y='valor2', data=data)
plt.title('valor2 por categoría', fontsize=16, fontweight="bold")
plt.show()
`````

**Pregunta**

¿Por qué es útil el boxen plot para valor2 y qué información entrega respecto a las colas de la distribución en cada categoría?

# Visualización 3: KDE plot (curva de densidad)

`````python
plt.figure(figsize=(8, 4))
sns.kdeplot(data['valor1'], label='valor1', fill=True)
sns.kdeplot(data['valor2'], label='valor2', fill=True)
plt.title('Densidad de valor1 y valor2', fontsize=16, fontweight="bold")
plt.legend()
plt.show()
`````

**Pregunta**

En el KDE plot de valor1 y valor2, ¿qué variable muestra mayor variabilidad y cómo se identifica en la gráfica?

# Visualización 4: Heatmap de correlación

`````python
plt.figure(figsize=(6, 5))
sns.heatmap(data[['valor1', 'valor2', 'valor3', 'valor4']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlaciones entre variables numéricas', fontsize=16, fontweight="bold")
plt.show()
`````

**Pregunta:**

¿Qué pares de variables presentan una correlación más alta según el heatmap? ¿Es positiva o negativa?

# Visualización 5: Pairplot (scatter matrix + distribuciones)

`````python
sns.pairplot(data, hue='categoria')
plt.suptitle('Relaciones entre variables', y=1.02)
plt.show()
`````

**Pregunta**

# Visualización 6: Strip plot (dispersión categórica con jitter)

En el contexto de un strip plot, el `jitter` se refiere a un pequeño desplazamiento aleatorio que se aplica a los puntos en la dirección perpendicular a la variable categórica (o a lo largo del eje que concentra los datos).

**¿Por qué se usa jitter?**

+ En un strip plot sin jitter, si muchos puntos tienen el mismo valor para la variable numérica y pertenecen a la misma categoría, se superpondrán unos sobre otros. Esto hace que no sea posible ver cuántos puntos hay realmente en ese lugar.
+ El jitter dispersa ligeramente esos puntos de forma horizontal (o vertical, según el gráfico) para que todos los puntos sean visibles y el gráfico sea más informativo.

`````python
plt.figure(figsize=(8, 4))
sns.stripplot(x='categoria', y='valor3', data=data, jitter=True)
plt.title('Dispersión de valor3 por categoría', fontsize=16, fontweight="bold")
plt.show()
`````

# Visualización 7: Count plot (conteo de categorías)

`````python
plt.figure(figsize=(6, 4))
sns.countplot(x='categoria', data=data)
plt.title('Frecuencia de cada categoría', fontsize=16, fontweight="bold")
plt.show()
`````

**Pregunta**

# Visualización 8: Histograma acumulativo en matplotlib

`````python
plt.figure(figsize=(8, 4))
plt.hist(data['valor4'], bins=20, cumulative=True, color='skyblue', edgecolor='black')
plt.title('Histograma acumulativo: valor4', fontsize=16, fontweight="bold")
plt.xlabel('valor4', fontsize=14, fontweight="bold")
plt.ylabel('Frecuencia acumulada', fontsize=14, fontweight="bold")
plt.show()
`````

**Pregunta**

¿Qué indica la pendiente del histograma acumulativo de valor4?
