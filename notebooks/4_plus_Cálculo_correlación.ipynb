# Ejemplo 1

## Conjunto de datos

`````python
# Creación de un DataFrame ficticio
data = {
    "Edad": [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    "Ingresos": [2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500],
    "HorasTrabajo": [40, 45, 50, 55, 60, 35, 30, 25, 20, 15],
    "SatisfaccionTrabajo": ["Bajo", "Medio", "Alto", "Alto", "Alto", "Bajo", "Bajo", "Medio", "Alto", "Medio"]
}

data_frame = pd.DataFrame(data)
`````

## Cálculo

`````python
# Convertir la variable objetivo en numérica
satisfaction_mapping = {"Bajo": 1, "Medio": 2, "Alto": 3}
data_frame["Satisfaccion"] = data_frame["SatisfaccionTrabajo"].map(satisfaction_mapping)

# Cálculo de las correlaciones
correlaciones = data_frame.corr(numeric_only=True)["Satisfaccion"].drop("Satisfaccion")
print("Correlaciones con la variable objetivo:")
print(correlaciones)

# Visualización de las correlaciones
plt.figure(figsize=(3, 3))
sns.heatmap(correlaciones.to_frame().T, annot=True, cmap="coolwarm",  linecolor="black")
plt.title("Correlación con la variable objetivo", fontsize=16, fontweight="bold")
plt.show()
`````

---
##Interpretaciones

---
Edad:

+ La correlación es muy baja y positiva.
+ Esto sugiere que la edad tiene una relación débil y directa con la satisfacción laboral. Sin embargo, es imposible concluir que a medida que aumenta la edad, aumente significativamente la satisfacción laboral.

Ingresos:

+ Similar a la correlación con la edad, es muy baja y positiva.
+ Esto implica que los ingresos tienen poca influencia directa en los niveles de satisfacción laboral. Las personas con ingresos más altos pueden tener una leve tendencia a reportar mayor satisfacción, pero la relación no es significativa.

Horas de Trabajo:

+ Esta correlación es moderada y positiva.
+ Sugiere que hay una relación más notable entre el número de horas trabajadas y la satisfacción laboral. Las personas que trabajan más horas parecen reportar mayores niveles de satisfacción laboral.
+ Sin embargo, dado que la correlación es menor que 0.5, no es una relación fuerte, y otros factores probablemente también influyan en la satisfacción laboral.

# Ejemplo 2

## Conjunto de datos

`````python
# Creación de un DataFrame ficticio
data = {
    "Edad": [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    "Ingresos": [2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500],
    "HorasTrabajo": [40, 45, 50, 55, 60, 35, 30, 25, 20, 15],
    "Promocion": ["No", "Si", "No", "Si", "Si", "No", "No", "Si", "Si", "No"]
}

data_frame = pd.DataFrame(data)
`````

## Cálculo

`````python
# Convertir la variable objetivo en numérica
data_frame["Promocion_c"] = data_frame["Promocion"].map({"No": 0, "Si": 1})

# Cálculo de las correlaciones
correlaciones = data_frame.corr(numeric_only=True)["Promocion_c"].drop("Promocion_c")
print("-"*60,"\nCorrelaciones con la variable objetivo (Promocion):\n","-"*60)
print(correlaciones)

# Visualización de las correlaciones
plt.figure(figsize=(3, 4))
sns.heatmap(correlaciones.to_frame().T, annot=True, cmap="coolwarm", cbar=False)
plt.title("Correlación de las variables con la variable objetivo", fontsize=16, fontweight="bold")
plt.show()
`````

# Ejemplo 3

## Conjunto de datos

`````python
# Amplia el DataFrame
data = {
    "Edad": [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    "Ingresos": [2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500],
    "HorasTrabajo": [40, 45, 50, 55, 60, 35, 30, 25, 20, 15],
    "Satisfaccion": [1, 2, 2, 3, 3, 2, 1, 3, 2, 1],
    "BalanceVidaTrabajo": [3, 4, 3, 5, 4, 2, 1, 5, 4, 3]
}

data_frame = pd.DataFrame(data)
`````

## Cálculo

`````python
# Calcula las correlaciones ampliadas
correlaciones = data_frame.corr(numeric_only=True)["Satisfaccion"].drop("Satisfaccion").round(3)
print("-"*40,"\nCorrelaciones ampliadas con Satisfacción\n","-"*40)
print(correlaciones)

# Visualiza las correlaciones
plt.figure(figsize=(8, 6))
sns.heatmap(correlaciones.to_frame().T, annot=True, cmap="coolwarm", cbar=False)
plt.title("Correlaciones con Satisfacción (ampliadas)", fontsize=16)
plt.show()
`````

---
## Interpretaciones

---
Edad:
+ Correlación débil y negativa.
+ Indica que la edad no tiene una influencia significativa en la satisfacción laboral.
+ La correlación negativa sugiere que un pequeño aumento en la edad podría estar asociado con una leve disminución en la satisfacción laboral, pero no es un factor determinante.

Ingresos:
+ Correlación débil y negativa.
+ Similar a la edad, los ingresos tienen una relación insignificante con la satisfacción laboral.
+ Contrario a lo esperado, mayores ingresos no parecen traducirse en mayor satisfacción laboral en este conjunto de datos.

Horas de Trabajo:
+ Correlación moderada y positiva.
+ Indica que trabajar más horas se asocia con un nivel más alto de satisfacción laboral.
+ Sin embargo, como la correlación no es extremadamente alta, este resultado podría estar influenciado por otros factores.

Balance Vida-Trabajo:
+ Correlación fuerte y positiva.
+ Esto significa que un mejor equilibrio entre vida personal y laboral está altamente relacionado con mayores niveles de satisfacción laboral.
+ Este es el factor más relevante en este análisis y sugiere que iniciativas para mejorar este balance podrían tener un impacto significativo en la satisfacción laboral.

**Conclusiones Generales**
+ El balance vida-trabajo es el factor más influyente en la satisfacción laboral, con una correlación fuerte. Las empresas deben priorizar políticas que favorezcan este aspecto, como horarios flexibles o trabajo remoto.

+ Horas de trabajo también tienen un impacto, aunque es menor que el balance vida-trabajo. Esto puede deberse a que trabajar más horas se percibe como compromiso laboral o acceso a mayores responsabilidades.

+ Edad e ingresos no tienen un impacto relevante en la satisfacción laboral en este conjunto de datos. Estos factores no deben considerarse prioritarios al diseñar estrategias para mejorar la satisfacción.
